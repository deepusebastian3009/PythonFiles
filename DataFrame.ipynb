```python
Filling Missing Values
```

```python
import pandas as pd
weather_data = {
    'day': ['2017-01-01','2017-01-04','	2017-01-05', '2017-01-06', '2017-01-07', '2017-01-08', 
            '2017-01-09', '2017-01-10', '2017-01-11'],
    'temperature': [32, 'NaN', 28, 'NaN', 32, 'NaN', 'NaN', 34, 40],
    'windspeed': [6, 7, 'NaN', 7, 'NaN', 'NaN', 'NaN', 8, 12],
    'event': ['Rain', 'Sunny', 'Snow', 'Nan', 'Rain', 'Sunny', 'NaN', 'Cloudy', 'Sunny']
}
df = pd.DataFrame(weather_data)
df
```

```python
type(df.day[2])
```

```python
type(df.windspeed[2])
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df=df.fillna(method='bfill')
new_df
```

```python
new_df=df.fillna(method='bfill', axis='columns')
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time") 
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
